```python
import cartopy.crs as ccrs #imports the crs module from the cartopy library which provides tools to work with different maps
import matplotlib.pyplot as plt #imports the pyplot module from matplotlib library which provides tools to plot data

fig = plt.figure(figsize=(10, 5)) #creating the plot and specifying it's size
ax = fig.add_subplot(1, 1, 1, projection=ccrs.PlateCarree()) #creating a subplot within our matplotlib figure and creating a map
ax.coastlines() #adding coastlines to the map we created earlier
plt.title('Map of the Earth') #adding plot title
plt.show() #showing the plot
```

In the following cell, add an arrow saying "We are here" that is pointing to Charlotte North Carolina


```python
import cartopy.crs as ccrs
import matplotlib.pyplot as plt

# Charlotte's coordinates
charlotte_lon = -80.8431
charlotte_lat = 35.2271

fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(1, 1, 1, projection=ccrs.PlateCarree())
ax.coastlines()

# Add the arrow and text
ax.annotate('We are here', xy=(charlotte_lon, charlotte_lat), xytext=(charlotte_lon + 10, charlotte_lat + 5),
            arrowprops=dict(facecolor='black', arrowstyle='->'),
            transform=ccrs.PlateCarree())

plt.title('Map of the Earth')
plt.show()
```

Now add major ocean labels

```python
import cartopy.crs as ccrs
import matplotlib.pyplot as plt

# Charlotte's coordinates
charlotte_lon = -80.8431
charlotte_lat = 35.2271

fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(1, 1, 1, projection=ccrs.PlateCarree())
ax.coastlines()

# Add the arrow and text
ax.annotate('We are here', xy=(charlotte_lon, charlotte_lat), xytext=(charlotte_lon + 10, charlotte_lat + 5),
            arrowprops=dict(facecolor='black', arrowstyle='->'),
            transform=ccrs.PlateCarree())
# Add major ocean labels (Atlantic, Pacific, etc.)
ax.text(-50, 0, 'Atlantic Ocean', transform=ccrs.PlateCarree())
ax.text(142, 15, 'Pacific Ocean', transform=ccrs.PlateCarree())
ax.text(-150, 0, 'Pacific Ocean', transform=ccrs.PlateCarree())
ax.text(-10, 72, 'Arctic Ocean', transform=ccrs.PlateCarree())
ax.text(0, -60, 'Southern Ocean', transform=ccrs.PlateCarree())
ax.text(60, -20, 'Indian Ocean', transform=ccrs.PlateCarree())

plt.title('Map of the Earth')
plt.show()
```
